```python
s = nvidia_all["memory_config"].astype(str).str.upper()
# Extract all GB numbers (supports "12 GB / 8 GB ...")
nvidia_all["vram_gb"] = (
    s.str.findall(r"(\d+)\s*GB")
     .apply(lambda x: "/".join(x) if len(x) > 0 else pd.NA)
)

# Extract memory type (GDDR6, GDDR6X, GDDR7, HBM etc.)
nvidia_all["memory_type"] = s.str.extract(r"(GDDR\dX?|HBM\d?)", expand=False)
```
